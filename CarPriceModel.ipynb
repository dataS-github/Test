```python
from sklearn.preprocessing import MinMaxScaler
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical = ['fuel', 'seller_type']
numeric_features = [col for col in df.columns if col not in categorical]

column_transformer = ColumnTransformer([
    ('ohe', OneHotEncoder(drop='first'), categorical),
    ('scaling', MinMaxScaler(), numeric_features)
])

df_transformed = column_transformer.fit_transform(df)
```

```python
df_transformed
```

```python
lst = list(column_transformer.transformers_[0][1].get_feature_names())
lst.extend(numeric_features)

df_transformed = pd.DataFrame(df_transformed, columns=lst)

df_transformed.head()
```

```python
X = df_transformed.drop(['selling_price'], axis=1)

y = df_transformed['selling_price']
```

```python
df_transformed['selling_price']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)

pred = model.predict(X_test)
```

```python
print(y_test)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
mean_absolute_error(y_test, pred), mean_squared_error(y_test, pred)
```
